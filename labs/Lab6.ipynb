## 1. Dimensionality Reduction

### 1.1 Principal Component Analysis (PCA)

We will start by implementing PCA for dimensionality reduction. We will also explore different parameters and visualize the results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar(label='Class')
plt.show()

# Scree plot
explained_variance = pca.explained_variance_ratio_
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

### Questions for Exploration

1. How does changing the number of principal components (`n_components`) affect the explained variance ratio?
Changing the number of principal components from 2 to 4, shows a deecrease the EVR per component (since the total variance is being spread out), adding more components doesn’t improve variance capture significantly beyond the first one.

2. What happens to the visualization when you use more or fewer principal components?
adding more principal components confirms most variance is in PC1
fewer components captures major variance and it is good for visualization
3. How does the choice of dataset affect the PCA results and their interpretation?
The choice of dataset greatly influences PCA results, especially in terms of dimensionality, feature correlation, and data scaling. PCA performs best on high-dimensional, correlated, and standardized data, but can be distorted by noise, outliers, or non-linear structures. Since PCA is unsupervised, it may miss class-specific patterns unless variance aligns with class differences, so careful preprocessing and dataset understanding are key for accurate interpretation.


### 1.2 t-Distributed Stochastic Neighbor Embedding (t-SNE)

Next, we will implement t-SNE for dimensionality reduction. We will also explore different parameters and visualize the results.

```python
from sklearn.manifold import TSNE

# Perform t-SNE with different perplexity values
perplexities = [5, 10, 30]  #we can change these values to see how they affect the visualization
fig, axes = plt.subplots(1, len(perplexities), figsize=(18, 6))

for i, perplexity in enumerate(perplexities):
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    X_tsne = tsne.fit_transform(X)
    axes[i].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
    axes[i].set_title(f't-SNE (perplexity={perplexity})')
    axes[i].set_xlabel('t-SNE Component 1')
    axes[i].set_ylabel('t-SNE Component 2')

plt.tight_layout()
plt.show()
```

### Questions for Exploration

1. How does changing the perplexity parameter affect the t-SNE visualization?
Increasing perplexity: emphasizes global structure, smoother transitions between clusters. A high perplexity Can help reveal overall group trends, but may lose detail.
Decreasing perplexity: emphasizes local structure, reveals smaller or subtle subgroups.
2. What happens to the visualization when you change the learning rate?
Changing the learning rate in t-SNE affects how the algorithm updates the point positions during optimization — and it directly influences the clarity and quality of your visualization.
Too low: slow convergence, poor separation.
Too high: unstable and noisy results.
Just right: preserves local/global structure, stable and interpretable clusters.
3. How does the choice of dataset affect the t-SNE results and their interpretation?
-Because t-SNE is nonlinear and unsupervised, its output is sensitive to both data preprocessing (like scaling and normalization) and dataset characteristics, so meaningful interpretation depends on a good understanding of the data's nature and context.
-If the dataset contains noise, outliers, or irrelevant features, t-SNE can produce misleading visualizations thus showing false clusters or obscuring true structure.




### 1.3 Uniform Manifold Approximation and Projection (UMAP)

Finally, we will implement UMAP for dimensionality reduction. We will also explore different parameters and visualize the results.

```python
import umap

# Perform UMAP with different n_neighbors values
n_neighbors_values = [5, 15, 50]
fig, axes = plt.subplots(1, len(n_neighbors_values), figsize=(18, 6))

for i, n_neighbors in enumerate(n_neighbors_values):
    umap_reducer = umap.UMAP(n_components=2, n_neighbors=n_neighbors, random_state=42)
    X_umap = umap_reducer.fit_transform(X)
    axes[i].scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis')
    axes[i].set_title(f'UMAP (n_neighbors={n_neighbors})')
    axes[i].set_xlabel('UMAP Component 1')
    axes[i].set_ylabel('UMAP Component 2')

plt.tight_layout()
plt.show()
```

### Questions for Exploration

1. How does changing the number of neighbors (`n_neighbors`) affect the UMAP visualization?
Small n_neighbors (e.g., 5) emphasizes local structure, capturing fine details and outliers but may fragment larger clusters and obscure global relationships. Medium n_neighbors (e.g., 15) balances local and global structure, offering clearer cluster transitions and is often a good default. Large n_neighbors (e.g., 50) highlights broad patterns and simplifies the visualization but can merge small clusters and lose fine-grained details.

2. What happens to the visualization when you change the minimum distance (`min_dist`)?

Small min_dist (0.01–0.1) creates tight, dense clusters with sharp boundaries, ideal for detecting outliers but risks over-segmenting data. Medium min_dist (0.1–0.5, default) balances cluster separation and smooth transitions, making it suitable for general-purpose analysis. Large min_dist (0.5–1.0) spreads points uniformly, emphasizing global topology at the cost of fine details, and its effect interacts with n_neighbors—smaller values preserve local structure while larger ones highlight broad patterns.


3. How does the choice of dataset affect the UMAP results and their interpretation?
The choice of dataset significantly impacts UMAP results, as the algorithm’s ability to reveal meaningful structure depends on the data’s inherent dimensionality, noise levels, and cluster separability—high-dimensional, noisy, or overlapping data may yield less distinct or misleading visualizations. Interpretation must therefore account for dataset-specific traits, such as feature scaling, sparsity, and domain context, to avoid overfitting artifacts or overlooking subtle patterns.

## 2. Visualization and Interpretation

We will visualize and interpret the results of different dimensionality reduction techniques. We will also explore different visualization techniques.

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Perform t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Perform UMAP
umap_reducer = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_reducer.fit_transform(X)

# Plot the results
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
axes[0].set_title('PCA')
axes[0].set_xlabel('Component 1')
axes[0].set_ylabel('Component 2')

axes[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
axes[1].set_title('t-SNE')
axes[1].set_xlabel('Component 1')
axes[1].set_ylabel('Component 2')

axes[2].scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis')
axes[2].set_title('UMAP')
axes[2].set_xlabel('Component 1')
axes[2].set_ylabel('Component 2')

plt.tight_layout()
plt.show()
```
